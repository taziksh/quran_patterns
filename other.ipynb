```python
import pandas as pd
import re

# 1) Load CSV data (adjust filename & column names to your dataset).
df = pd.read_csv("quran_data.csv")
print(df.head())  # quick look at the first few rows
```

```python
import re

# Function to remove diacritics and normalize text
def normalize_arabic(text):
    # Remove diacritics (harakaat)
    diacritic_pattern = re.compile(r'[\u0617-\u061A\u064B-\u0652\u0657-\u065B\u06D6-\u06ED]')
    text = re.sub(diacritic_pattern, '', text)
    
    # Normalize Alifs (أ, إ, آ, ٱ → ا)
    text = re.sub(r'[أإآٱ]', 'ا', text)
    
    # Optionally normalize Taa Marbuta (ة → ه) if relevant
    # text = text.replace('ة', 'ه')
    
    return text

# Apply normalization to the "ayah_ar" column
df["ayah_normalized"] = df["ayah_ar"].apply(normalize_arabic)

# Example: Search for a concept
def search_concept(word, df, column="ayah_normalized"):
    # Normalize the search word
    word_normalized = normalize_arabic(word)
    
    # Check which verses contain the word
    mask = df[column].str.contains(word_normalized, na=False)
    matches = df[mask]
    return matches

# Example Usage: Find verses with the concept "الدنيا"
concept = "الدنيا"
results = search_concept(concept, df)
print(f"Found {len(results)} verses containing the concept '{concept}'.")
results[["surah_no", "ayah_no_surah", "ayah_ar"]].head()
```
